```python
import pandas as pd


url="https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
```

```python
df=pd.read_csv(url,header=None)

df.head(2)
```

```python
heading=[]      # making some random heading for ther header in df
count=0

line1="a b c d e f"
line2="g h i j k l"

list1=line1.split()
list2=line2.split()


for i in range (1,25):
    for j in list1:
        for k in list2:
            
            word=j+k
            
            count+=1
            
            if count <=26:
                heading.append(word) 
```

```python
print(heading)
```

```python
df.columns=heading # adding the heading
```

### Drop missing values in python 

```python
df1=df.dropna()                          # use dropna() method to remove missing values row's
df1
```

```python
df2=df.dropna(axis=1)             # use axis =1 for dropping the column which missing value present
df2  
```

```python
df1=df.dropna(axis=0) # use axis =0 for dropping the column which missing value is present
df1
```
